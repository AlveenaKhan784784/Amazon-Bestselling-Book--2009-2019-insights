```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['Price','User Rating']])
plt.show()
```

- Pirce may vary

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Genre', y='Price', data=df)
plt.title("Price Distribution by Genre")
plt.xlabel("Genre")
plt.ylabel("Price")
plt.show()
```

```python
top_books = df.sort_values(by='User Rating',ascending=False).head(10)
print(top_books[['Name','Author','User Rating']])
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="User Rating", y="Name", data=top_books, palette="viridis")

plt.title("Top 10 Books by User Rating")
plt.xlabel("User Rating")
plt.ylabel("Book Name")
plt.xlim(4, 5)  # Since ratings are usually between 4-5
plt.show()
```
